```python
#Load the 'tips' dataset from Seaborn and display the first 5 rows.
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
tips.head(5)
```

```python
#Create a scatter plot of total_bill vs tip using the 'tips' dataset.
sns.scatterplot(x='tip',y='total_bill',data=tips)
plt.title("Total_bill vs Tip")
plt.show()
```

```python
#Plot a line graph for total_bill vs tip using Seaborn
sns.lineplot(x='tip',y='total_bill',data=tips)
plt.title("Total_bill vs Tip")
plt.show()
```

```python
# Draw a bar plot to show the average tip for each day in the 'tips' dataset
import pandas as pd
avg_tips=tips.groupby('day')['tip'].mean()
avg_tips=pd.DataFrame(avg_tips).reset_index()
sns.barplot(x='day',y='tip',data=avg_tips)
plt.title("Average Tip by Day")
plt.show()
```

```python
# Use a count plot to display how many records are available for each day
sns.countplot(x='day',data=tips)
plt.title("Count of Records by Day")
plt.show()
```

```python
# Make a box plot to show total_bill distribution across different days.
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
# Create a violin plot for total_bill vs day.
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
# Plot a histogram of the 'total_bill' column with KDE using histplot.
sns.histplot(x='total_bill',data=tips,bins=20,kde=True)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
# Load the 'iris' dataset and draw a pairplot for all features.
iris=sns.load_dataset('iris')
sns.pairplot(iris)
plt.show()
```

```python
# Use the 'titanic' dataset and show a heatmap of its numeric correlation matrix.
titanic=sns.load_dataset('titanic')
corr=titanic.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Numeric Correlation Matrix")
plt.show()
```

```python
# Use the 'penguins' dataset and draw a scatter plot of bill_length_mm vs
# flipper_length_mm.
penguins=sns.load_dataset('penguins')
sns.scatterplot(x='flipper_length_mm',y='bill_length_mm',data=penguins)
plt.title("Bill Length vs Flipper Length")
plt.show()
```

```python
# Plot a box plot for 'fare' against 'class' using the 'titanic' dataset
sns.boxplot(x='class',y='fare',data=titanic)
plt.title("Fare by Class")
plt.show()
```

```python
# Create a count plot to show the number of survivors and non-survivors in the Titanic
# dataset.
titanic.head(5)
sns.countplot(x='survived',data=titanic)
plt.title("Survived vs Not survived")
plt.show()
```

```python
# Create a heatmap of the correlation matrix from the 'penguins' dataset using
# numeric_only=True.
corr=penguins.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Numeric Correlation Matrix")
plt.show()
```

```python
# Load the 'diamonds' dataset and draw a violin plot of price by cut.
diamond=sns.load_dataset('diamonds')
sns.violinplot(x='cut',y='price',data=diamond)
plt.title("Price by Cut")
plt.show()
```

```python
# Draw a pairplot using the 'penguins' dataset with hue as 'species'
sns.pairplot(penguins, hue='species')
plt.show()
```

```python
# Apply a dark theme to your plots using set_style and recreate a scatter plot.
sns.set_style('darkgrid')
sns.scatterplot(x='flipper_length_mm',y='bill_length_mm',data=penguins)
plt.title("Bill Length vs Flipper Length")
plt.show()
```

```python
# Convert all categorical columns in the 'tips' dataset to numeric and create a full
# correlation heatmap.
tips_copy=tips.copy()
tips_copy['sex'] = tips['sex'].map({'Female': 0, 'Male': 1})
tips_copy['smoker'] = tips['smoker'].map({'No': 0, 'Yes': 1})
tips_copy['day'] = tips['day'].map({'Thur': 0, 'Fri': 1, 'Sat': 2, 'Sun': 3})
tips_copy['time'] = tips['time'].map({'Lunch': 0, 'Dinner': 1})
corr=tips_copy.corr()
print(corr)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Numeric Correlation Matrix")
plt.show()
```
